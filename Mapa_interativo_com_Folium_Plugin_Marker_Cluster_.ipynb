#Mapa interativo com a biblioteca Folium do Python

##Conhecendo o plugin Marker Cluster do Folium

Documentação da bibioteca: https://python-visualization.github.io/folium/plugins.html

image.png

```python
# instalar biblioteca
!pip install folium

#importar biblioteca
import folium
from folium.plugins import MarkerCluster
#from folium.plugins import BeautifyIcon
```

```python
# definindo dimensões de exibição da figura 
fig = folium.Figure(width = 800, height = 400) 

# criar mapa interativo
m = folium.Map(location=[-10, -50], # centro do mapa
               zoom_start=3, # nivel de zoom inicial
               ).add_to(fig)
```

```python
# criar o grupo agregado de marcadores "mcg"
mcg = folium.plugins.MarkerCluster(control=False)
m.add_child(mcg)
```

```python
# criar o grupo 1
g1 = folium.plugins.FeatureGroupSubGroup(mcg, 'g1')
m.add_child(g1)

# criar o grupo 2
g2 = folium.plugins.FeatureGroupSubGroup(mcg, 'g2')
m.add_child(g2)

# criar o grupo 3
g3 = folium.plugins.FeatureGroupSubGroup(mcg, 'g3')
m.add_child(g3)

# criar o grupo 4
g4 = folium.plugins.FeatureGroupSubGroup(mcg, 'g4')
m.add_child(g4)
```

```python
# add coordenadas no grupo 1 "g1"
folium.Marker([-20,-40]).add_to(g1)
folium.Marker([-21,-41]).add_to(g1)

# add coordenadas no grupo 2 "g2"
folium.Marker([-10,-50]).add_to(g2)
folium.Marker([-11,-51]).add_to(g2)

# add coordenadas no grupo 3 "g3"
folium.Marker([-5,-40]).add_to(g3)
folium.Marker([-6,-41]).add_to(g3)

# add coordenadas no grupo 4 "g4"
folium.Marker([-7,-55]).add_to(g4)
folium.Marker([-11,-56]).add_to(g4)
```

```python
# add layer control
folium.LayerControl().add_to(m)
```
